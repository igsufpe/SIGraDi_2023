#Setup

```python
import pandas as pd
```

#Dataset import (.csv)

```python
train_df = pd.read_csv('/content/sample_data/DATASET-TPS-SIGRADI-R00-TREINO.csv',encoding="ISO-8859-1", delimiter=';')
valid_df = pd.read_csv('/content/sample_data/DATASET-TPS-SIGRADI-R00-TESTE.csv',encoding="ISO-8859-1", delimiter=';')
```

#Data visualization

```python
train_df.head()
```

#Separating the response variable

```python
y_train = train_df['status']
y_valid = valid_df['status']
del train_df['status']
del valid_df['status']
```

```python
y_train[0:10]
```

```python
train_df.head()
```

#Creating training /test set

```python
x_train = train_df.values
x_valid = valid_df.values
```

```python
x_train[0]
```

# Summarizing the Training and Validation Data

```python
x_train.shape
```

```python
y_train.shape
```

#Normalize x_train and x_valid

```python
x_train = x_train / 54
x_valid = x_valid / 54
```

```python
x_train.min()
```

```python
x_train.max()
```

#Import tensorflow.keras

```python
import tensorflow.keras as keras
num_classes = 2
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_valid = keras.utils.to_categorical(y_valid, num_classes)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units = 512, activation='relu', input_shape=(30,)))
model.add(Dense(units = 512, activation='relu'))
model.add(Dense(units = num_classes, activation='softmax'))
```

## Summarizing the Model

```python
model.summary()
```

## Compiling the Model

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

##Train the Model

```python
model.fit(x_train, y_train, epochs=50, verbose=1, validation_data=(x_valid, y_valid))
```
